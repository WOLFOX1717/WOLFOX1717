```python
# Tratamiento de datos
# ==============================================================================
import pandas as pd
import numpy as np
from sklearn.datasets import load_diabetes

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
from matplotlib import style
import seaborn as sns

# Preprocesado y análisis
# ==============================================================================
import statsmodels.api as sm
import pingouin as pg
from scipy import stats
from scipy.stats import pearsonr

# Configuración matplotlib
# ==============================================================================
plt.style.use('ggplot')

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
url = ('https://raw.githubusercontent.com/JoaquinAmatRodrigo/' +
       'Estadistica-machine-learning-python/master/data/Howell1.csv')
datos = pd.read_csv(url)

# Se utilizan únicamente información de individuos mayores de 18 años.
datos = datos[datos.age > 18]

datos.info()
```

```python
# Gráfico
# ==============================================================================
fig, ax = plt.subplots(1, 1, figsize=(6,4))
ax.scatter(x=datos.height, y=datos.weight, alpha= 0.8)
ax.set_xlabel('Altura')
ax.set_ylabel('Peso');
```

```python
# Gráfico distribución variables
# ==============================================================================
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))

axs[0].hist(x=datos.height, bins=20, color="#3182bd", alpha=0.5)
axs[0].plot(datos.height, np.full_like(datos.height, -0.01), '|k', markeredgewidth=1)
axs[0].set_title('Distribución altura (height)')
axs[0].set_xlabel('height')
axs[0].set_ylabel('counts')

axs[1].hist(x=datos.weight, bins=20, color="#3182bd", alpha=0.5)
axs[1].plot(datos.weight, np.full_like(datos.weight, -0.01), '|k', markeredgewidth=1)
axs[1].set_title('Distribución peso (weight)')
axs[1].set_xlabel('weight')
axs[1].set_ylabel('counts')


plt.tight_layout();
```

```python
# Gráfico Q-Q
# ==============================================================================
fig, axs = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

sm.qqplot(
    datos.height,
    fit   = True,
    line  = 'q',
    alpha = 0.4,
    lw    = 2,
    ax    = axs[0],

)
for line in axs[0].get_lines():
    line.set_color('blue') #cambia de color la linea

axs[0].set_title('Gráfico Q-Q height', fontsize = 10, fontweight = "bold")
axs[0].tick_params(labelsize = 7)

sm.qqplot(
    datos.height,
    fit   = True,
    line  = 'q',
    alpha = 0.4,
    lw    = 2,
    ax    = axs[1]
)
for line in axs[1].get_lines():
    line.set_color('purple')
axs[1].set_title('Gráfico Q-Q height', fontsize = 10, fontweight = "bold")
axs[1].tick_params(labelsize = 7)
```

```python
# Normalidad de los residuos Shapiro-Wilk test
# ==============================================================================
shapiro_test = stats.shapiro(datos.height)
print(f"Variable height: {shapiro_test}")
shapiro_test = stats.shapiro(datos.weight)
print(f"Variable weight: {shapiro_test}")
```

```python
# Normalidad de los residuos D'Agostino's K-squared test
# ==============================================================================
k2, p_value = stats.normaltest(datos.height)
print(f"Variable height: Estadítico = {k2}, p-value = {p_value}")
k2, p_value = stats.normaltest(datos.weight)
print(f"Variable weight: Estadítico = {k2}, p-value = {p_value}")
```

```python
# Transformación logarítmica de los datos
# ==============================================================================
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 4))

sm.qqplot(
    np.log(datos.height),
    fit   = True,
    line  = 'q',
    alpha = 0.4,
    lw    = 2,
    ax    = ax
)
for line in ax.get_lines():
    line.set_color('purple')
ax.set_title('Gráfico Q-Q log(height)', fontsize = 13)
ax.tick_params(labelsize = 7)


shapiro_test = stats.shapiro(np.log(datos.height))
print(f"Variable height: {shapiro_test}")
```

```python
# Cálculo de correlación con Pandas
# ==============================================================================
print('Correlación Pearson: ', datos['weight'].corr(datos['height'], method='pearson'))
print('Correlación spearman: ', datos['weight'].corr(datos['height'], method='spearman'))
print('Correlación kendall: ', datos['weight'].corr(datos['height'], method='kendall'))
```

```python
# Cálculo de correlación y significancia con Scipy
# ==============================================================================
r, p = stats.pearsonr(datos['weight'], datos['height'])
print(f"Correlación Pearson: r={r}, p-value={p}")

r, p = stats.spearmanr(datos['weight'], datos['height'])
print(f"Correlación Spearman: r={r}, p-value={p}")

r, p = stats.kendalltau(datos['weight'], datos['height'])
print(f"Correlación Pearson: r={r}, p-value={p}")
```

```python
# Cálculo de correlación, significancia e intervalos con pingouin
# ==============================================================================
display(pg.corr(datos['weight'], datos['height'], method='pearson'))
display(pg.corr(datos['weight'], datos['height'], method='spearman'))
display(pg.corr(datos['weight'], datos['height'], method='kendall'))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

# Descomponer la serie temporal
from statsmodels.tsa.seasonal import seasonal_decompose

# Assuming your NDVI data is in the 'Spring' column, replace 'Spring' if necessary
# The period should be adjusted based on the actual seasonality in your data.
# Since you have 10 observations and need at least 2 cycles,
# You might consider a smaller period if your data has a shorter seasonal cycle.
# For example, if your data has a yearly seasonality, but you only have 10 years of data,
# you can't use period=12 (monthly). Instead, try period=1 (yearly).
decomposition = seasonal_decompose(df['Spring'], model='additive', period=1)  # Adjust period to 1

# ... (rest of your plotting code remains the same)
# Graficar la descomposición
plt.figure(figsize=(14, 8))
plt.subplot(411)
# Changed 'NDVI' to 'Spring' to match the actual column name in the DataFrame
plt.plot(df['Spring'], label='Serie original')
plt.title('Serie temporal original')
plt.subplot(412)
plt.plot(decomposition.trend, label='Tendencia', color='orange')
plt.title('Tendencia')
plt.subplot(413)
plt.plot(decomposition.seasonal, label='Estacionalidad', color='green')
plt.title('Estacionalidad')
plt.subplot(414)
plt.plot(decomposition.resid, label='Residuales', color='red')
plt.title('Residuales')
plt.tight_layout()
plt.show()
```

```python
# 1. Instalar las bibliotecas necesarias
!pip install pandas matplotlib statsmodels

# 2. Importar las bibliotecas necesarias
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# 3. Cargar los datos (ajusta la ruta de tu archivo CSV)
df = pd.read_csv('/content/NDVI_tent.csv') # Reemplaza con la ruta correcta

# 4. Convertir la columna de fecha a tipo datetime
df['Year'] = pd.to_datetime(df['Year'], format='%Y') # Reemplaza 'fecha' con el nombre de tu columna de fecha

# 5. Establecer la columna de fecha como índice
df.set_index('Year', inplace=True) # Reemplaza 'fecha' con el nombre de tu columna de fecha

# 6. Graficar la serie temporal de NDVI
plt.figure(figsize=(12, 6))
# Reemplaza 'NDVI' con el nombre de tu columna de NDVI, por ejemplo, 'Spring'
plt.plot(df['Spring'], label='NDVI')
plt.title('Serie temporal de NDVI')
plt.xlabel('Fecha')
plt.ylabel('NDVI')
plt.legend()
plt.grid(True)
plt.show()

# 7. Realizar la diferenciación si es necesario (diferenciación de orden 1)
# Reemplaza 'NDVI' con el nombre de tu columna de NDVI
df['NDVI_diff'] = df['Spring'].diff().dropna()

# 8. Graficar el ACF y PACF para elegir p y q
plt.figure(figsize=(12, 6))
# Adjust lags to be less than or equal to the length of your differenced data
# Since your differenced data has 9 observations, set lags to a maximum of 9.
# You can use a smaller number like 5 if you only need to see the first few lags.
plot_acf(df['NDVI_diff'].dropna(), lags=4)  # ACF (autocorrelación)
plt.show()


# 9. Ajustar el modelo ARIMA
# Aquí seleccionas p, d, q según los gráficos ACF y PACF
# En este caso, como ejemplo p=1, d=1, q=1 (ajusta según tu análisis)
# Reemplaza 'NDVI' con el nombre de tu columna de NDVI
modelo_arima = ARIMA(df['Spring'], order=(1, 1, 1))  # Cambia los valores de p, d, q si es necesario
modelo_resultado = modelo_arima.fit()

# 10. Resumen del modelo
print(modelo_resultado.summary())

# 11. Graficar los resultados
plt.figure(figsize=(12, 6))
# Reemplaza 'NDVI' con el nombre de tu columna de NDVI
plt.plot(df['Spring'], label='Serie original')
plt.plot(modelo_resultado.fittedvalues, label='Ajuste ARIMA', color='orange', linestyle='--')
plt.title('Serie temporal de NDVI con ajuste ARIMA')
plt.xlabel('Fecha')
plt.ylabel('NDVI')
plt.legend()
plt.grid(True)
plt.show()

# 12. Hacer predicciones
# Predicción de los siguientes 12 pasos (ajusta el número de pasos según tus necesidades)
# predicciones = modelo_resultado.predict(start=len(df), end=len(df) + 11) # Ajusta el rango de predicción
# plt.plot(predicciones, label='Predicciones', color='red', linestyle='--')
# plt.show()
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook  # Fixed indentation
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper
import pandas as pd
from bokeh.palettes import Category10


output_notebook()

# Load and preprocess your data here - removed extra spaces
df = pd.read_csv('/content/NDVI_tent.csv')
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.set_index('Year', inplace=True)
source = ColumnDataSource(df)

# Create the figure - removed extra spaces
p = figure(title="NDVI Trend Over Time", x_axis_type="datetime", width=800, height=400)

seasons = ['Spring', 'Summer', 'Autumn', 'Winter']
colors = Category10[len(seasons)]

for i, season in enumerate(seasons):
    p.line(x='Year', y=season, source=source, line_width=2, line_color=colors[i], legend_label=season, name=season)
    p.circle(x='Year', y=season, source=source, size=8, fill_color=colors[i], legend_label=season, name=season)

# Apply styling attributes - removed extra spaces
p.background_fill_color = '#2F2F2F'
p.border_fill_color = '#2F2F2F'
p.outline_line_color = '#444444'

p.xaxis.axis_line_color = "white"
p.yaxis.axis_line_color = "white"
p.xaxis.axis_label_text_color = "white"
p.yaxis.axis_label_text_color = "white"
p.xaxis.major_label_text_color = "white"
p.yaxis.major_label_text_color = "white"
p.xaxis.major_tick_line_color = "white"
p.yaxis.major_tick_line_color = "white"
p.xaxis.minor_tick_line_color = "white"
p.yaxis.minor_tick_line_color = "white"

p.xgrid.grid_line_dash = [6, 4]
p.ygrid.grid_line_dash = [6, 4]
p.xgrid.grid_line_alpha = 0.3
p.ygrid.grid_line_alpha = 0.3

p.title.text_color = "white"

hover = HoverTool(tooltips=[("Year", "@Year{%Y}"), ("NDVI", "@{$name}"), ("Season", "$name")],
                 formatters={'@Year': 'datetime', '@{$name}': 'printf'})
p.add_tools(hover)

p.legend.location = "top_right"
p.legend.background_fill_alpha = 0.5
p.legend.label_text_color = "white"


show(p)
```
